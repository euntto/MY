```python
# 보라 / 11월 1일 월요일 / 19:50 마감 
close = [342, 343, 342, 342, 340, 342, 341, 339, 340, 338, 337, 339, 340, 339, 340, 341, 341, 338, 336, 336
         ,337, 337, 337, 339, 340, 340, 341, 341, 344, 344, 344, 350, 349, 343, 345, 346, 346, 346, 346, 347]
close = pd.DataFrame(close,columns = ['Close'])
```

```python
def TSI(data, n_days1 = 10, n_days2 = 5, signal = 5):
    delta = data['Close'].diff()

    mom1 = delta.ewm(span=n_days1).mean().ewm(span=n_days2).mean()
    mom2 = delta.abs().ewm(span=n_days1).mean().ewm(span=n_days2).mean()
    # tsi
    data['tsi'] = mom1/mom2
    #EMA
    data['signal'] = data['tsi'].ewm(span=signal).mean()

    return data[['tsi','signal']]
```
