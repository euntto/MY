```python
baseline_feature = ['ID', 'AGE', 'EVENT_YN','SEX']
lab_feature = ['ID', 'VISIT','GLU',  'A1C','GFR']


# In[167]:


def MERGE_TIME(data_baseline, data_lab, time):
    merged_data = pd.merge(data_baseline[baseline_feature], data_lab[(data_lab['VISIT']==time)][lab_feature], how='inner',left_on='ID',right_on = 'ID')
    return merged_data

```
